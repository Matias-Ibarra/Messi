```python
'''
Añado los últimos goles a nivel club, marcados con el Inter de Miami, para hacer un análisis estadístico general de
el rendimiento goleador de Messi.
'''
```

```python
import pandas as pd
df_con_ccaa = pd.read_csv('data/messi_goles_con_caa.csv')
df_con_ccaa.head()
```

```python
#Las Columnas 'Matchday' y 'Playing_Position' no van a ser parte del análisis.
df_con_ccaa.drop(['Matchday', 'Playing_Position'], axis=1, inplace=True)
df_con_ccaa.head()
```

```python
goles_inter_miami = pd.read_csv('data/goles_inter_miami.csv', sep=';')
goles_inter_miami.head()
```

```python
# Arreglo los datos de las distintas columnas para que coincidan
# Primero la columna 'Date' en df_con_ccaa
df_con_ccaa['Date'] = pd.to_datetime(df_con_ccaa['Date'])
print("Tipo de dato Date:", df_con_ccaa['Date'].dtype)
df_con_ccaa['Date'] = df_con_ccaa['Date'].dt.strftime('%d-%m-%Y')
df_con_ccaa.head()
```

```python
# Ahora en goles_inter_miami
goles_inter_miami['Date'] = pd.to_datetime(goles_inter_miami['Date'])
print("Tipo de dato Date:", goles_inter_miami['Date'].dtype)
goles_inter_miami['Date'] = goles_inter_miami['Date'].dt.strftime('%d-%m-%Y')
goles_inter_miami.head()
```

```python
# Ahora la columna 'Result'
df_con_ccaa['Result'] = df_con_ccaa['Result'].str.replace(":", "-")
df_con_ccaa.head()
```

```python
# Hay valores AET que quiero eliminar
df_con_ccaa['Result'].unique()
```

```python
df_con_ccaa['Result'] = df_con_ccaa['Result'].str.replace("AET", "")
```

```python
# Además voy a separar cada valor en columnas distintas y convertirlas en un entero.
df_con_ccaa['Score home'] = df_con_ccaa['Result'].str.split('-').str[0].astype(int)
df_con_ccaa['Score visitant'] = df_con_ccaa['Result'].str.split('-').str[1].astype(int)
df_con_ccaa.drop(['Result'], axis=1, inplace=True)
df_con_ccaa.head()
```

```python
# Quiero que At_score tenga formato, 2-0, 3-1, etc.
df_con_ccaa['At_score'] = df_con_ccaa['At_score'].str.replace(":", "-")
```

```python
df_con_ccaa.head()
```

```python
goles_inter_miami['Result'].unique()
```

```python
# Ahora en goles_inter_miami
goles_inter_miami['Score home'] = goles_inter_miami['Result'].str.split('-').str[0].astype(int)
goles_inter_miami['Score visitant'] = goles_inter_miami['Result'].str.split('-').str[1].astype(int)
goles_inter_miami.drop(['Result'], axis=1, inplace=True)
goles_inter_miami.head()
```

```python
# Agrego los goles de Inter de Miami al total de goles.
goles_europa_eeuu = pd.concat([df_con_ccaa, goles_inter_miami], axis = 0).reset_index(drop=True)
goles_europa_eeuu
```
