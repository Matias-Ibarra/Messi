```python
'''
Uso de web scrapping para extraer datos. En este proyecto se extrajeron datos de la web neogol para obtener el listado de los 50 máximos
goleadores históricos, y de wikipedia la altura de c/u para hacer una comparativa.

'''
```

```python
from bs4 import BeautifulSoup as bs
import requests
from fake_useragent import UserAgent
ua = UserAgent()
url = 'https://www.neogol.com/maximos-goleadores-de-la-historia-del-futbol/'
headers = {'User-Agent': ua.random}
response = requests.get(url, headers=headers)
```

```python
print(response)
```

```python
html = response.content
```

```python
print(html)
soup = bs(html, 'html.parser')
print(soup)
```

```python
# Extraigo los nombres de los goleadores

todo = soup.find('table').find_all('td', style='width: 30.4581%;')
jugadores = []
for i in todo:
    #print(i.get_text())
    jugadores.append(i.get_text().lstrip())

print(jugadores)
```

```python
# Extraigo los goles de c/u de los goleadores

todo = soup.find('table').find_all('td', height='20')
goles = []
for i in todo:
    #print(i.get_text())
    goles.append(i.get_text())

print(goles)
```

```python
# Extraigo los partidos de c/u de los goleadores

partidos = []
# El primer valor tiene otro formato
primer_valor = soup.find('table').find_all('td', style='width: 14.2104%; padding-bottom: 0.2em; padding-top: 0.2em; text-align: center;')
for i in primer_valor:
    partidos.append(i.get_text())

todo = soup.find('table').find_all('td', style='padding-bottom: 0.2em; padding-top: 0.2em; width: 14.2104%; text-align: center;')
for i in todo:
    #print(i.get_text())
    partidos.append(i.get_text())

print(partidos)
```

```python
# Extraigo el promedio de c/u de los goleadores
promedio = []
# El primer valor tiene otro formato
primer_valor = soup.find('table').find_all('td', style='width: 11.0841%; padding-bottom: 0.2em; padding-top: 0.2em; text-align: center;')
for i in primer_valor:
    promedio.append(i.get_text())

todo = soup.find('table').find_all('td', style='padding-bottom: 0.2em; padding-top: 0.2em; width: 11.0841%; text-align: center;')
for i in todo:
    #print(i.get_text())
    promedio.append(i.get_text())

print(promedio)
```

```python
# Creo un diccionario
tabla_goleadores = {"Jugador": jugadores,
                    "Goles": goles,
                    "Partidos": partidos,
                    "Promedio": promedio
}
```

```python
# Lo convierto en dataframe
import pandas as pd
goleadores = pd.DataFrame(tabla_goleadores)
goleadores.index = goleadores.index + 1
goleadores
```

```python
import numpy as np
altura = []
for i in jugadores:
    ua = UserAgent()
    url = 'https://es.wikipedia.org/wiki/'+i.replace(" ", "_")
    headers = {'User-Agent': ua.random}
    response = requests.get(url, headers=headers)
    html = response.content
    soup = bs(html, 'html.parser')
    altura_individual = soup.find('th', text='Altura')
    if altura_individual:
        # Accede al valor de la altura que está en el siguiente <td>
        altura_valor = altura_individual.find_next('td').text
        #print("Altura:", altura_valor)
        altura.append((altura_valor[1:]))
    else:
        altura.append(np.nan)
        #print("Altura no encontrada")

altura
```
